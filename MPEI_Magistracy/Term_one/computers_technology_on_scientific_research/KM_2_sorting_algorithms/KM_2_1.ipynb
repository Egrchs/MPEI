```python
#1)   Напишите   и   протестируйте   код   на   языке   Python   для   следующих вариантов сортировки:
#   а) Сортировка вставками

def insertion_sort(A):                                                    # Алгоритм сортировки вставками
    for i in range(1, len(A)):
        temp = A[i]
        j = i - 1
        while (j >= 0 and temp < A[j]):
            A[j + 1] = A[j]
            j = j - 1
        A[j + 1] = temp
    return A    



A = [random.randint(0,99 ) for i in range(20)]                             # Создание массива
print('Исходный массив :')
print(A,"\n")                                                              # Вывод исходного массива
insertion_sort(A)                                                          # Сортировка массива
print('Отсортированный массив :')
print(A)                                                                   # Вывод отсортированного массива

t1=np.zeros(800)                                                           # Создание массива для результатов обработки времени
t11=np.zeros(800)
n1=np.arange(1,801,1)                                                      # Создание массива длин вектора 
n=1
while n<801:
    v = [random.randint(0, 9) for i in range(n)]                            # Создание вектора размера n 
    t11[n-1]=1.5e-8*n**2
    r = timeit.timeit('insertion_sort(v)', globals=globals(), number=2)     # Расчет времени выполнения алгоритма 
    t1[n-1]=r/2                                                             # Расчет среднего времени 
    n+=1

fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                          # Построение графиков 
plt.grid(visible=True)
ax.set_xlabel("Длина массива") 
ax.set_ylabel("Время выполнения сек") 
ax.set_title('Оценка времени выполнения алгоритма сортировки вставками ') 
ax.plot(n1,t1,n1,t11)
ax.legend(labels = ("Экспирементальная временная зависимость","Теоретическая временная зависимость"), loc="upper left");
```

```python
# б) Сортировка выбором.

def selection_sort(B):                                                    # Алгоритм сортировки выбором
    for i in range(0, len(B) - 1):
        smallest = i
        for j in range(i + 1, len(B)):
            if B[j] < B[smallest]:
                smallest = j
        B[i], B[smallest] = B[smallest], B[i]

B = [random.randint(0, 99) for i in range(20)]                             # Создание массива
print('Исходный массив :')
print(B,"\n")                                                              # Вывод исходного массива
selection_sort(B)                                                          # Сортировка массива
print('Отсортированный массив :')
print(B)  

t2=np.zeros(800)                                                           # Создание массива для результатов обработки времени 
t22=np.zeros(800)
n2=np.arange(1,801,1)                                                      # Создание массива длин вектора 
n=1
while n<801:
    v = [random.randint(0, 9) for i in range(n)]                            # Создание вектора размера n 
    t22[n-1]=2.5e-8*n**2.01
    r = timeit.timeit('selection_sort(v)', globals=globals(), number=5)     # Расчет времени выполнения алгоритма 
    t2[n-1]=r/5                                                             # Расчет среднего времени 
    n+=1
   

fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                          # Построение графиков 
plt.grid(visible=True)
ax.set_xlabel("Длина массива") 
ax.set_ylabel("Время выполнения сек") 
ax.set_title('Оценка времени выполнения алгоритма сортировки выбором ') 
ax.plot(n2,t2,n2,t22);
ax.legend(labels = ("Экспирементальная временная зависимость","Теоретическая временная зависимость"), loc="upper left");
```

```python
# в) Пирамидальная сортировка.
def heapsort(alist):                                                             # Алгоритм пирамидальной сортировки
    build_max_heap(alist)
    for i in range(len(alist) - 1, 0, -1):
        alist[0], alist[i] = alist[i], alist[0]
        max_heapify(alist, index=0, size=i)
 
def parent(i):
    return (i - 1)//2
 
def left(i):
    return 2*i + 1
 
def right(i):
    return 2*i + 2
 
def build_max_heap(alist):
    length = len(alist)
    start = parent(length - 1)
    while start >= 0:
        max_heapify(alist, index=start, size=length)
        start = start - 1
 
def max_heapify(alist, index, size):
    l = left(index)
    r = right(index)
    if (l < size and alist[l] > alist[index]):
        largest = l
    else:
        largest = index
    if (r < size and alist[r] > alist[largest]):
        largest = r
    if (largest != index):
        alist[largest], alist[index] = alist[index], alist[largest]
        max_heapify(alist, largest, size)

C = [random.randint(0, 99) for i in range(20)]                              # Создание массива
print('Исходный массив :')
print(C,"\n")                                                               # Вывод исходного массива
heapsort(C)                                                                 #  Сортировка массива
print('Отсортированный массив :')
print(C) 

t3=np.zeros(800)                                                           # Создание массива для результатов обработки времени 
t33=np.zeros(800)
n3=np.arange(1,801,1)                                                      # Создание массива длин вектора 
n=1
while n<801:
    v = [random.randint(0, 9) for i in range(n)]                            # Создание вектора размера n 
    t33[n-1]=n*np.log(n)*6e-7
    r = timeit.timeit('heapsort(v)', globals=globals(), number=5)           # Расчет времени выполнения алгоритма 
    t3[n-1]=r/5                                                             # Расчет среднего времени 
    n+=1
   

fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                          # Построение графиков 
plt.grid(visible=True)
ax.set_xlabel("Длина массива") 
ax.set_ylabel("Время выполнения сек") 
ax.set_title('Оценка времени выполнения алгоритма пирамидальной сортировки ') 
ax.plot(n3,t3,n3,t33);
ax.legend(labels = ("Экспирементальная временная зависимость","Теоретическая временная зависимость"), loc="upper left");
```

```python
# г) Сортировка разбиением / слиянием.
def merge_sort(nums):                                                   # Алгоритм разбиением слиянием
    if len(nums) > 1: 
        mid = len(nums)//2
        left = nums[:mid] 
        right = nums[mid:]
        merge_sort(left) 
        merge_sort(right) 
        i = j = k = 0
        while i < len(left) and j < len(right): 
            if left[i] < right[j]: 
                nums[k] = left[i] 
                i+=1
            else: 
                nums[k] = right[j] 
                j+=1
            k+=1
        while i < len(left): 
            nums[k] = left[i] 
            i+=1
            k+=1
        while j < len(right): 
            nums[k] = right[j] 
            j+=1
            k+=1
D = [random.randint(0, 99) for i in range(20)]                             # Создание массива
print('Исходный массив :') 
print(D,"\n")                                                              # Вывод исходного массива
merge_sort(D)                                                              # Сортировка массива
print('Отсортированный массив :')
print(D)       


t4=np.zeros(800)                                                           # Создание массива для результатов обработки времени 
t44=np.zeros(800)
n4=np.arange(1,801,1)                                                      # Создание массива длин вектора 
n=1
while n<801:
    v = [random.randint(0, 9) for i in range(n)]                            # Создание вектора размера n 
    t44[n-1]=n*np.log(n)*3.5e-7
    r = timeit.timeit('merge_sort(v)', globals=globals(), number=5)         # Расчет времени выполнения алгоритма 
    t4[n-1]=r/5                                                             # Расчет среднего времени 
    n+=1
   

fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                          # Построение графиков 
plt.grid(visible=True)
ax.set_xlabel("Длина массива") 
ax.set_ylabel("Время сек") 
ax.set_title('Оценка времени выполнения алгоритма сортировки слиянием ') 
ax.plot(n4,t4,n4,t44);
ax.legend(labels = ("Экспирементальная временная зависимость","Теоретическая временная зависимость"), loc="upper left");
```

```python
# д) Быстрая сортировка по Хоару

def QuickSort(A):                                                         # Сортировка по Хоару
    if len(A) <= 1:
        return A
    else:
        q = random.choice(A)
        L = []
        M = []
        R = []
        for elem in A:
            if elem < q:
                L.append(elem) 
            elif elem > q: 
                R.append(elem) 
            else: 
                M.append(elem)
        return QuickSort(L) + M + QuickSort(R)
    
E = [random.randint(0, 99) for i in range(20)]                              # Создание массива
print('Исходный массив :')
print(E,"\n")                                                               # Вывод исходного массива
E1=QuickSort(E)                                                             # Сортировка массива
print('Отсортированный массив :')
print(E1)  


t5=np.zeros(800)                                                           # Создание массива для результатов обработки времени 
t55=np.zeros(800)  
n5=np.arange(1,801,1)                                                      # Создание массива длин вектора 
n=1
while n<801:
    v = [random.randint(0, 9) for i in range(n)]                            # Создание вектора размера n 
    t55[n-1]=n*np.log(n)*5e-8
    r = timeit.timeit('QuickSort(v)', globals=globals(), number=5)          # Расчет времени выполнения алгоритма 
    t5[n-1]=r/5                                                             # Расчет среднего времени 
    n+=1
   

fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                          # Построение графиков 
plt.grid(visible=True)
ax.set_xlabel("Длина массива") 
ax.set_ylabel("Время выполнения сек") 
ax.set_title('Оценка времени выполнения алгоритма сортировки по Хоару') 
ax.plot(n5,t5,n5,t55);
ax.legend(labels = ("Экспирементальная временная зависимость","Теоретическая временная зависимость"), loc="upper left");
```

```python
# е) Блочная сортировка через хеш-функцию «x/3».
def bucket_sort(array):                                              # Алгоритм сортировки через хэш-функцию x/3
    bucket = []
    
    
    for i in range(round(max(array)/3)+1):
        bucket.append([])

    
    for j in array:
        index_b = round(j/3)
        bucket[index_b].append(j)

    
    for i in range((round(max(array)/3)+1)):
        bucket[i] = sorted(bucket[i])

    
    k = 0
    for i in range((round(max(array)/3)+1)):
        for j in range(len(bucket[i])):
            array[k] = bucket[i][j]
            k += 1
    return array


F = [random.randint(0, 99) for i in range(20)]                             # Создание массива
print('Исходный массив :')
print(E,"\n")                                                              # Вывод исходного массива
F1=bucket_sort(F)                                                          # Сортировка массива
print('Отсортированный массив :')
print(F1)  

t6=np.zeros(800)                                                           # Создание массива для результатов обработки времени 
t66=np.zeros(800)
n6=np.arange(1,801,1)                                                      # Создание массива длин вектора 
n=1
while n<801:
    v = [random.randint(0, 9) for i in range(n)]                            # Создание вектора размера n 
    t66[n-1]=n*np.log(n)*5e-8
    r = timeit.timeit('bucket_sort(v)', globals=globals(), number=5)        # Расчет времени выполнения алгоритма 
    t6[n-1]=r/5                                                             # Расчет среднего времени 
    n+=1
   

fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                          # Построение графиков 
plt.grid(visible=True)
ax.set_xlabel("Длина массива") 
ax.set_ylabel("Время выполнения сек") 
ax.set_title('Оценка времени выполнения алгоритма сортировки через хэш-функцию ') 
ax.plot(n6,t6,n6,t66);
ax.legend(labels = ("Экспирементальная временная зависимость","Теоретическая временная зависимость"), loc="upper left");
```

```python
# 3) Подготовьте анимацию процесса сортировки, выполняемой в п. 2


array = [5, 7, 7, 4, 6, 5, 3, 9, 7, 7, 7, 6, 0, 2, 4, 1, 2, 8, 6, 3, 7, 6]
def selection_sort(a):
   
    N = len(a)
    for i in range(N - 1):
        m = a[i]
        p = i
        for j in range(i + 1, N):
            if m > a[j]:
                m = a[j]
                p = j

        if p != i:
            t = a[i]
            a[i] = a[p]
            a[p] = t
        yield (a, i, p)
    #print(a)       


 


def barlist_step():
  return next(arr) # итерируемся по сортировке

fig = plt.figure()
# инициализируем генератор
arr = selection_sort(array)
# инициализируем бары barlist_step()[0] = массив 
barcollection = plt.bar(range(len(array)), barlist_step()[0], color='gray')

def animate(current_frame):
    print(current_frame)
    try:
        # получаем значения сортировки и два индекса перестановки
        y, c1, c2 = barlist_step()
        for i, b in enumerate(barcollection):
            b.set_height(y[i])
            if (i == c1 or i == c2): 
                # подсвечиваем выбранные индексы
                b.set_color('green') 
            else:
                b.set_color('grey') 
    except StopIteration: # итерации закончились
        pass # можно вывести количество итераций           

anim = animation.FuncAnimation(
    fig,
    animate,
    frames=100, # количество кадров (зависит от алгоритма)
    repeat=False,
    interval=1 # задает интервал и fps для видео
    
)


anim.save('sort_2_1.gif')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
import random
# Функция анимации 
def animate(_list, generator, title, save, save_frames=1000):
    n = len(_list)
    fig, ax = plt.subplots()
    ax.set_title(title)
 
    bar_rects = ax.bar(range(len(_list)), _list, align="edge")  
    ax.set_xlim(0, n)
    ax.set_ylim(0, int(1.1*n))
    text = ax.text(0.01, 0.95, "", transform = ax.transAxes)
    iteration = [0]
    def _animate(array, rects, iteration):
        for rect, val in zip(rects, array):
            rect.set_height(val)
        iteration[0] += 1
        text.set_text(f"iterations: {iteration[0]}")
   
    anim = FuncAnimation(fig, func=_animate,
        fargs=(bar_rects, iteration), frames=generator, interval=200,
        repeat=False, save_count=save_frames)
    if save:
        writer = PillowWriter(fps=60)
        anim.save(title + ".gif", writer=writer)
    plt.show()	
 
def quicksort(array, start, end):
    if start >= end:
        return
    pivot = array[start]
    swap = start
    for i in range(start+1, end+1):
        if array[i] <= pivot:
            swap += 1
            array[swap], array[i] = array[i], array[swap]
        yield array
    array[start], array[swap],  = array[swap], array[start]
    yield array
 
    yield from quicksort(array, start, swap-1)
    yield from quicksort(array, swap+1, end)
 

array = [2, 2, 4, 2, 3, 3, 3, 3, 3, 4, 2, 6, 6, 6, 6, 6, 6, 7, 7, 8, 10, 12]
random.shuffle(array)
animate(array, quicksort(array, 0, 21), "quicksort", True)
```
