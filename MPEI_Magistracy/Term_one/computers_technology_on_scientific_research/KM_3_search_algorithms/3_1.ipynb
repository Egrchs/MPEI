```python
# Подключение необходимых библиотек
import random
import matplotlib.pyplot as plt
import numpy as np
import timeit
import string
import pprint
import pandas as pd
from random import shuffle

sample_str = string.ascii_letters  
```

1)   Реализуйте   алгоритм   последовательного   поиска.   Определите
вычислительную   сложность   алгоритма   в   среднем   на   задаче   поиска   числа   в
массиве, буквы в строке.

```python

def linear_Search(A,  key):                                               # Алгоритм линейного поиска 
  for i in range(0, len(A)): 
    if (A[i] == key): 
        return i 
  return -1  

 

A = [random.randint(0,19 ) for i in range(20)]                             # создание массива случайных чисел
key1 = 7 
print("Массив :  ",A)
print("Ключ к массиву :",key1)

res1 = linear_Search(A,  key1)                                             # поиск числа в массиве
if(res1 == -1): 
    print("Элемент не найден") 
else: 
    print("Индекс найденого элемента: ", res1) 

B = ''.join(random.choice(sample_str) for _ in range(20))                  # создание строки случайных символов
key11 = 'u' 
print("Строка :  ",B)
print("Ключ к строке :",key11)

res2 = linear_Search(B,  key11)                                            # поиск символа в строке
if(res2 == -1): 
    print("Элемент не найден") 
else: 
    print("Индекс найденого элемента: ", res2)   


t1=np.zeros(50)                                                             # Создание массива для результатов обработки времени
t11=np.zeros(50)
n1=np.arange(1,51,1)                                                        # Создание массива длин вектора 
n=1
while n<51:
    A = [random.randint(0, 9) for i in range(n)]                             # Создание вектора размера n 
    t1[n-1] = timeit.timeit('linear_Search(A,key1)', globals=globals())     # Расчет времени выполнения алгоритма 
    B = ''.join(random.choice(sample_str) for _ in range(n) )                # Создание вектора размера n 
    t11[n-1] = timeit.timeit('linear_Search(B,key11)', globals=globals())   # Расчет времени выполнения алгоритма 
    n+=1

fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                          # Построение графиков 
plt.grid(visible=True)
ax.set_xlabel("Длина массива") 
ax.set_ylabel("Время выполнения сек") 
ax.set_title('Оценка времени выполнения алгоритма последовательного поиска ') 
ax.plot(n1,t1,n1,t11)
ax.legend(labels = ("поиск числав в массиве","Поиск символа в строке"), loc="upper left");
```

2) Повторите п. 1 для алгоритма бинарного поиска.

```python
def binary_Search(A, key):                                                # Алгоритм бинарного поиска
    sorted_list=sorted(A)
    low = 0
    high = len(sorted_list) - 1

    while low <= high:
        mid = (low + high) // 2
        guess = sorted_list[mid]
        if guess == key:
            return mid
        if guess > key:
            high = mid - 1
        else:
            low = mid + 1
    return -1


C = [random.randint(0,19 ) for i in range(20)]                             # создание массива случайных чисел
key2 = 3 
print("Массив :  ",C)
print("Ключ к массиву :",key2)

res1 = linear_Search(C,  key2)                                             # поиск числа в массиве
if(res1 == -1): 
    print("Элемент не найден") 
else: 
    print("Индекс найденного элемента: ", res1) 

D = ''.join(random.choice(sample_str) for _ in range(20))                  # создание строки случайных символов
key22 = 'z' 
print("Строка :  ",D)
print("Ключ к строке :",key22)

res2 = linear_Search(D,  key22)                                            # поиск символа в строке
if(res2 == -1): 
    print("Элемент не найден") 
else: 
    print("Индекс найденного элемента: ", res2)   


t2=np.zeros(50)                                                             # Создание массива для результатов обработки времени
t22=np.zeros(50)
n2=np.arange(1,51,1)                                                        # Создание массива длин вектора 
n=1
while n<51:
    C = [random.randint(0, 9) for i in range(n)]                            # Создание вектора размера n 
    t2[n-1] = timeit.timeit('binary_Search(C,key2)', globals=globals())     # Расчет времени выполнения алгоритма 
    D = ''.join(random.choice(sample_str) for _ in range(n))                # Создание вектора размера n 
    t22[n-1] = timeit.timeit('binary_Search(D,key22)', globals=globals())   # Расчет времени выполнения алгоритма 
    n+=1

fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                          # Построение графиков 
plt.grid(visible=True)
ax.set_xlabel("Длина массива") 
ax.set_ylabel("Время выполнения сек") 
ax.set_title('Оценка времени выполнения алгоритма бинарного поиска ') 
ax.plot(n2,t2,n2,t22)
ax.legend(labels = ("поиск числав в массиве","Поиск символа в строке"), loc="upper left");
```

3)   Напишите   функцию   хеширования   для   массива,   каждый   элемент
которого — строка. Можно сделать любой «простой» хеш (деление на простое
число, полином, и т. п.).

```python
def hash8(message, table,tab_size):                    # Алгоритм Пирсона
    hash = len(message) % tab_size
    for i in message:
        hash = table[(hash+ord(i)) % tab_size]
    return hash
```
