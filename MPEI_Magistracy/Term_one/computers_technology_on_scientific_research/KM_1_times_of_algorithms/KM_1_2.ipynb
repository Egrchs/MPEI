```python
# Подключение необходимых библиотек
import random
import matplotlib.pyplot as plt
import numpy as np
import timeit
```

```python
#1.1) Сгенерируйте n-мерный случайный вектор v = [v1, v2, ..., vn] с неотрицательными элементами. Для полученного вектора v осуществите
# реализацию алгоритмов:а) Любой алгоритм, имеющий константную сложность.
  
t1=np.zeros(2000)                                                           # Создание массива для результатов обработки врмени 
n1=np.arange(1,2001,1)                                                      # Создание массива длин вектора 
n=1
while n<2001:
    v = [random.randint(0, 9) for i in range(n)]                            # Создание вектора размера n 
    r = timeit.timeit('v.append(2)', globals=globals(), number=5)           # Расчет времени выполнения алгоритма 
    t1[n-1]=r/5                                                             # Расчет среднего времени 
    n+=1

fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                          # Построение графиков 
plt.grid(visible=True)
ax.set_xlabel("Длина массива") 
ax.set_ylabel("Время выполнения сек") 
ax.set_title('Оценка времени выполнения алгоритма константной сложности ') 
ax.plot(n1,t1);
```

```python
 #1.2) Алгоритм, выполняющий сумму элементов.
def summa( L=[]):                                                           # Функция подсчета суммы
    i=0
    s=0
    n=len(L)
    while i<n:
        s=s+L[i]
        i+=1
    return s

t2=np.zeros(2000)                                                           # Создание массива для результатов обработки врмени 
n2=np.arange(1,2001,1)                                                      # Создание массива длин вектора 
n=1
while n<2001:
    v = [random.randint(0, 9) for i in range(n)]                            # Создание вектора размера n 
    r = timeit.timeit('summa(v)', globals=globals(), number=5)              # Расчет времени выполнения алгоритма 
    t2[n-1]=r/5                                                             # Расчет среднего времени 
    n+=1
   

fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                          # Построение графиков 
plt.grid(visible=True)
ax.set_xlabel("Длина массива") 
ax.set_ylabel("Время выполнения сек") 
ax.set_title('Оценка времени выполнения алгоритма выполняющего сумму элементов ') 
ax.plot(n2,t2);
```

```python
 #1.3) Алгоритм, выполняющий произведение элементов.
def mult( L=[]):                                                             # Функция подсчета произведения
    i=0
    p=0
    n=len(L)
    while i<n:
        p=p*L[i]
        i+=1
    return p

t3=np.zeros(2000)                                                           # Создание массива для результатов обработки врмени 
n3=np.arange(1,2001,1)                                                      # Создание массива длин вектора 
n=1
while n<2001:
    v = [random.randint(0, 9) for i in range(n)]                            # Создание вектора размера n 
    r = timeit.timeit('mult(v)', globals=globals(), number=5)               # Расчет времени выполнения алгоритма 
    t3[n-1]=r/5                                                             # Расчет среднего времени 
    n+=1
   

fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                          # Построение графиков 
plt.grid(visible=True)
ax.set_xlabel("Длина массива") 
ax.set_ylabel("Время выполнения сек") 
ax.set_title('Оценка времени выполнения алгоритма выполняющего произведение элементов ') 
ax.plot(n3,t3);
```

```python
#1.4) Алгоритм, вычисляющий значение многочлена по классической схеме:
def pol( L=[]):                                                             # Функция вычисления значения многочлена по классической схеме
    i=0
    x=2
    pol=0
    n=len(L)
    while i<n:
        pol+=x*L[i]**(n-1)
        i+=1
    return pol

t4=np.zeros(2000)                                                           # Создание массива для результатов обработки врмени 
n4=np.arange(1,2001,1)                                                      # Создание массива длин вектора 
n=1
while n<2001:
    v = [random.randint(0, 9) for i in range(n)]                            # Создание вектора размера n 
    r = timeit.timeit('pol(v)', globals=globals(), number=5)                # Расчет времени выполнения алгоритма 
    t4[n-1]=r/5                                                             # Расчет среднего времени 
    n+=1
   

fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                          # Построение графиков 
plt.grid(visible=True)
ax.set_xlabel("Длина массива") 
ax.set_ylabel("Время сек") 
ax.set_title('Оценка времени выполнения алгоритма вычисления значения многочлена по классической схем ') 
ax.plot(n4,t4);
```

```python
#1.5) Алгоритм, вычисляющий значение многочлена по схеме Горнера:
def gorner( L=[]):                                                          # Функция вычисления значения многочлена по схеме Горнера                     
    x=2
    n=len(L)
    gor=L[n-1]
    while n>0:
        gor+=gor*x+L[n-2]
        n-=1
    return gor

t5=np.zeros(2000)                                                           # Создание массива для результатов обработки врмени 
n5=np.arange(1,2001,1)                                                      # Создание массива длин вектора 
n=1
while n<2001:
    v = [random.randint(0, 9) for i in range(n)]                            # Создание вектора размера n 
    r = timeit.timeit('gorner(v)', globals=globals(), number=5)             # Расчет времени выполнения алгоритма 
    t5[n-1]=r/5                                                             # Расчет среднего времени 
    n+=1
   

fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                          # Построение графиков 
plt.grid(visible=True)
ax.set_xlabel("Длина массива") 
ax.set_ylabel("Время сек") 
ax.set_title('Оценка времени выполнения алгоритма вычисления значения многочлена по схеме Горнера ') 
ax.plot(n5,t5,n5,t4);
ax.legend(labels = ("Схема Горнера","Классический метод"), loc="upper left");
```

```python
#2) Сгенерируйте случайные матрицы A и B размером n × n с неотрицательными элементами. Найдите обычное матричное произведение матриц A и B.

def mult_max(A=[],B=[]):                                                   # Функция вычисления произведения  
   C=[]
   C=A.dot(B)
   return C

t6=np.zeros(50)                                                             # Создание массива для результатов обработки врмени 
n6=np.arange(1,1001,20)                                                     # Создание массива размеров матриц

n=1
t=0
while n<1001:
    A = np.random.randint(10, size=(n, n))                                 # Создание квадратной матрицы А
    B = np.random.randint(10, size=(n, n))                                 # Создание квадратной матрицы В
    r= timeit.timeit('mult_max(A,B)', globals=globals(), number=5)         # вычисление времени выполнения алгоритма 
    t6[t]=r/5
    n+=20
    t+=1

fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                         # Создание графика 
plt.grid(visible=True)
ax.set_xlabel("Длина массива") 
ax.set_ylabel("Время сек") 
ax.set_title('Оценка времени выполнения алгоритма умножения матриц ') 
ax.plot(n6,t6);
```

```python
import string


def naive_pattern_search(s: str, pattern: str) -> list:
    """
    >>> naive_pattern_search("ABAAABCDBBABCDDEBCABC", "ABC")
    [4, 10, 18]
    >>> naive_pattern_search("ABC", "ABAAABCDBBABCDDEBCABC")
    []
    >>> naive_pattern_search("", "ABC")
    []
    >>> naive_pattern_search("TEST", "TEST")
    [0]
    >>> naive_pattern_search("ABCDEGFTEST", "TEST")
    [7]
    """
    pat_len = len(pattern)
    position = []
    for i in range(len(s) - pat_len + 1):
        match_found = True
        for j in range(pat_len):
            if s[i + j] != pattern[j]:
                match_found = False
                break
        if match_found:
            position.append(i)
    return position


if __name__ == "__main__":
    assert naive_pattern_search("ABCDEFG", "DE") == [3]
    print(naive_pattern_search("ABAAABCDBBABCDDEBCABC", "ABC"))


t7=np.zeros(4000)                                                           # Создание массива для результатов обработки врмени 
n7=np.arange(1,4001,1)                                                      # Создание массива длин вектора 
n=1
while n<4001:
    A =  ''.join(random.choice(string.ascii_lowercase) for i in range (n))                             # Создание вектора размера n 
    r = timeit.timeit('naive_pattern_search(A, "sdjjrt")', globals=globals(), number=3)        # Расчет времени выполнения алгоритма 
    t7[n-1]=r/3                                                            # Расчет среднего времени 
    n+=1
   

fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                          # Построение графиков 
plt.grid(visible=True)
ax.set_xlabel("Длина массива") 
ax.set_ylabel("Время сек") 
ax.set_title('Оценка времени выполнения алгоритма  ') 
ax.plot(n7,t7);
```
