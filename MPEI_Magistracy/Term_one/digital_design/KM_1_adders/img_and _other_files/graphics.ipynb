```python
import matplotlib.pyplot as plt
import numpy as np
```

# Serial adder

```python
W             = [8,       16,     32,     64,     128  ]                             # 
Freq_serial   = [110.12,  70.65,  56.45,  30.65,  15.17]                             #  serial_adder
Freq_lokahead = [129.94,  69.77,  54.9,   31.46,  15.1 ]                             #  carry_lokahead



fig, ax = plt.subplots(figsize = [12,6],dpi = 100)                                   #   
plt.grid(visible=True)  
ax.set_xlabel("") 
ax.set_ylabel(", ") 
ax.set_title('   ') 
ax.plot(W,Freq_serial, W, Freq_lokahead);
ax.legend(labels = ("Serial_adder","Carry_lokahead"), loc="upper right");
```
