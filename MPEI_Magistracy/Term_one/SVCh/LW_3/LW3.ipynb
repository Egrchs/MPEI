```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv('VAH.txt',sep='\s+', header=None)
data = pd.DataFrame(data)
x=data[0]
y=data[1]
z=data[3]

   
fig,ax1 = plt.subplots()
ax2 = ax1.twinx()    
plt.rcParams['figure.dpi'] = 250 
plt.rcParams["figure.figsize"] = (9.6,5.4)
plt.rcParams["axes.grid"] = True
 

ax1.set_ylabel(" I  А") 
ax1.plot(x,y*0.05) 
ax1.set_xlabel("U   В") 
ax2.set_ylabel(" P Вт",color = 'darkred') 

ax2.plot(x,z,color = 'darkred',linestyle='dashed') ;
ax1.legend(labels = ("Ток",), loc="upper left");
ax2.legend(labels = ("Мощность",), loc="upper right");
```
