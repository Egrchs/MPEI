```python
import matplotlib.pyplot as plt
import pandas as pd

```

```python
# CdSe Фр 117
data = pd.read_csv('CdSe.txt',sep='\s+', header=None)
data = pd.DataFrame(data)
x=data[0]
y=data[1]
Eg1=(3e8*6.6e-34)/(1.6e-19*(x[3]+x[8])*1e-9/2)
Eg1='%.2f'%Eg1
print("Ширина запрешённой зоны - ",Eg1,"эВ")
fig, ax = plt.subplots()        
plt.rcParams['figure.dpi'] = 250 
plt.rcParams["figure.figsize"] = (9.6,5.4)
plt.rcParams["axes.grid"] = True

ax.set_xlabel("λ,нм") 
ax.set_ylabel("I,мкА") 
ax.set_title('Спектр тока') 
ax.plot(x,y,) ;
```
