```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('R_r5_nl.txt',sep='\s+', header=None)
data = pd.DataFrame(data)
x=data[0]
y=data[1]
I=np.zeros(3)
for i in range(3):
     I[i]=y[i]/1000
Rt=(x[2]-x[0])/(I[2]-I[0])
Rt0=Rt/1e6
Rt0='%.2f'%Rt0
print("Темновое сопротивление",Rt0,"МОм")    
fig, ax = plt.subplots(figsize = [12,6],dpi = 90)        
plt.rcParams['figure.dpi'] = 250 
plt.rcParams["figure.figsize"] = (9.6,5.4)
plt.rcParams["axes.grid"] = True

ax.set_xlabel("U, В") 
ax.set_ylabel("I, А") 
ax.set_title('ВАХ фоторезистора(темновая r = 5см)') 
ax.plot(x,I) ;
```
