```python
import matplotlib.pyplot as plt
import pandas as pd

```

```python

data = pd.read_csv('spect.txt',sep='\s+', header=None)
data = pd.DataFrame(data)
x=data[0]
y=data[1]

 
fig, ax = plt.subplots()        
plt.rcParams['figure.dpi'] = 250 
plt.rcParams["figure.figsize"] = (9.6,5.4)
plt.rcParams["axes.grid"] = True

ax.set_xlabel("lambda,В") 
ax.set_ylabel("I,А") 
ax.set_title('') 
ax.plot(x,y) ;
```

```python
data1 = pd.read_csv('VAH.txt',sep='\s+', header=None)
data1 = pd.DataFrame(data1)
x1= data1[0]
y1= data1[1]
z1= data1[2]
fig, ax = plt.subplots()        
plt.rcParams['figure.dpi'] = 250 
plt.rcParams["figure.figsize"] = (9.6,5.4)
plt.rcParams["axes.grid"] = True

ax.set_xlabel(",") 
ax.set_ylabel("") 
ax.set_title('') 
ax.plot(z1,abs(x1*y1)) ;
```
